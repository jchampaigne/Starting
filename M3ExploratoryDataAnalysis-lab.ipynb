## Hands on Lab


Import the pandas module.


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Load the dataset into a dataframe.


```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")
```

## Distribution


### Determine how the data is distributed


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


Plot the distribution curve for the column `ConvertedComp`.


```python
# your code goes here
dfComp = df.dropna(subset=['ConvertedComp'])  #removes all empty rows within the ConvertedComp column, save as new dataframe
dfComp.ConvertedComp.isnull().sum()           #confirms no empty rows
sns.distplot(dfComp.ConvertedComp)            #plots the distribution curve for ConvertedComp
```

Plot the histogram for the column `ConvertedComp`.


```python
# your code goes here
dfComp['ConvertedComp'].plot(kind='hist')
plt.title=('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Count')
plt.show()
```

What is the median of the column `ConvertedComp`?


```python
# your code goes here
dfComp['ConvertedComp'].median()
```

How many responders identified themselves only as a **Man**?


```python
# your code goes here
dfComp['Gender'].value_counts()['Man']
```

Find out the  median ConvertedComp of responders identified themselves only as a **Woman**?


```python
# your code goes here
dfComp.loc[dfComp['Gender'] == 'Woman', ['ConvertedComp']].median()

#code below gives median of all groups
#dfComp.groupby('Gender')['ConvertedComp'].median()
```

Give the five number summary for the column `Age`?

```python
# your code goes here
dfComp['Age'].describe().loc[['min','25%','mean','75%','max']]
```

Plot a histogram of the column `Age`.

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
sns.boxplot(data=dfComp['ConvertedComp'])
plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = dfComp['ConvertedComp'].quantile(0.25)

Q3 = dfComp['ConvertedComp'].quantile(0.75)

IQR = Q3 - Q1

IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print('lower bounds: ' + str(lower_limit))
print('upper bounds: ' + str(upper_limit))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
((dfComp['ConvertedComp'] < (Q1 - 1.5 * IQR)) | (dfComp['ConvertedComp'] > (Q3 + 1.5 * IQR))).sum()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here

df_no_outliers = dfComp[(dfComp['ConvertedComp'] < upper_limit) & (dfComp['ConvertedComp'] > lower_limit)]
df_no_outliers
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# your code goes here
df_no_outliers.corr()['Age']
```
